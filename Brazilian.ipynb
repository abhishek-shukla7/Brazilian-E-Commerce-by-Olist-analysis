TASK 1.3 — Chain Join

orders
   ↓
customers
   ↓
order_items   ← (this sets the grain)
   ↓
products
   ↓
category_translation


Orders-> Customers

```python
df=orders.merge(
    customer,on="customer_id",
    how="left"

)

print("After orders + customers: ",df.shape)
```

```python
df=df.merge(
    order_items,
    on="order_id",
    how="left"
)
print("After adding order items: ",df.shape)
print("Unique orders:",df['order_id'].nunique())
```

```python
df=df.merge(
    product,
    on="product_id",
    how="left"
)

print("after adding the products: ",df.shape)
```

```python
df=df.merge(
    category_translation,
    on="product_category_name",
    how="left"
)

print("after adding the category translation: ",df.shape)
```

```python
print("Rows:",df.shape[0])
print("unique order_id: ",df['order_id'].nunique())
print("unique product_id: ",df['product_id'].nunique())


df[['order_id','product_id']].duplicated().sum()
```
