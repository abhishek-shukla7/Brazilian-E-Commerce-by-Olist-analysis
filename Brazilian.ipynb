DAY 3

TASK 3.1 — Create a Customer-Level Dataset

```python
rfm_base=df_day2[[
                 "customer_unique_id",
                 "order_id",
                 "order_purchase_timestamp",
                 "total_payment_value"
                 ]

].dropna(subset=["customer_unique_id"])
```

```python
customer_df=(
    rfm_base
    .groupby("customer_unique_id",as_index=False)
    .agg(
        last_purchase_date=("order_purchase_timestamp","max"),
        frequency=("order_id","nunique"),
        monetary=("total_payment_value","sum")
    )
)
```

```python
customer_df.shape
customer_df.head()
```

```python
customer_df["customer_unique_id"].duplicated().sum()
```

```python
rfm_customer_base=customer_df.copy()
```

```python
reference_date=rfm_customer_base["last_purchase_date"].max()
reference_date
```

```python
rfm_customer_base["recency"]=(
    reference_date - rfm_customer_base["last_purchase_date"]
).dt.days
```

```python
rfm_customer_base.rename(
    columns={
        "frequency": "frequency",
        "monetary": "monetary"
    },
    inplace=True
)
```

```python
rfm_customer_base[["recency", "frequency", "monetary"]].describe()
```

```python
rfm_customer_base.isna().sum()
```

```python
rfm = rfm_customer_base.copy()
rfm.head()
```

Task3.3:Quantile Scoring

```python
# Recency score (reverse: recent buyers get higher score)
rfm["R_score"] = pd.qcut(
    rfm["recency"],
    q=5,
    labels=[5, 4, 3, 2, 1]
).astype(int)

# Frequency score
rfm["F_score"] = pd.qcut(
    rfm["frequency"].rank(method="first"),
    q=5,
    labels=[1, 2, 3, 4, 5]
).astype(int)

# Monetary score
rfm["M_score"] = pd.qcut(
    rfm["monetary"],
    q=5,
    labels=[1, 2, 3, 4, 5]
).astype(int)
```

```python
rfm[["R_score", "F_score", "M_score"]].describe()
```

```python
rfm["R_score"].value_counts().sort_index()
rfm["F_score"].value_counts().sort_index()
rfm["M_score"].value_counts().sort_index()
```

```python
rfm["RFM_score"] = (
    rfm["R_score"].astype(str) +
    rfm["F_score"].astype(str) +
    rfm["M_score"].astype(str)
)
```

```python
rfm_scored = rfm.copy()
rfm_scored.head()
```

Customer Segmentation

```python
def rfm_segment(row):
    if row["R_score"] >= 4 and row["M_score"] >= 4:
        return "Champions"
    elif row["R_score"] >= 3 and row["M_score"] >= 4:
        return "Loyal Customers"
    elif row["R_score"] >= 4 and row["M_score"] <= 2:
        return "Potential Loyalists"
    elif row["R_score"] == 3 and row["M_score"] == 3:
        return "Need Attention"
    elif row["R_score"] <= 2 and row["M_score"] >= 3:
        return "At Risk"
    elif row["R_score"] <= 2 and row["M_score"] <= 2:
        return "Lost Customers"
    else:
        return "Others"
```

```python
rfm_scored["segment"] = rfm_scored.apply(rfm_segment, axis=1)
```

```python
rfm_scored["segment"].value_counts()
```

```python
segment_distribution = (
    rfm_scored["segment"]
    .value_counts(normalize=True)
    .mul(100)
    .round(2)
)

segment_distribution
```
